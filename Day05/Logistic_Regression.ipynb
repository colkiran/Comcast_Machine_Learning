```python
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
# Dataset
X = np.array([[29], [15], [33], [28], [39]])
y = np.array([0, 0, 1, 1, 1])
```

```python
# Train Logistic Regression Mode
model = LogisticRegression()
model.fit(X, y)

#1 Probability of passing for 33 hours
prob_33 = model.predict_proba([[36]])[0][1]
print("Probability of passing at 33 hours :", prob_33)
```

```python
# 2. Minimum hours for getting marks > 95%
hours_range = np.arange(10, 50).reshape(-1, 1)
probs = model.predict_proba(hours_range)[:, 1]

threshold_hours = hours_range[np.argmax(probs > 0.95)][0]
print("Minimum hours for marks > 95% probability :", threshold_hours)
```

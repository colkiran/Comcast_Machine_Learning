```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Dataset
X = np.array([
    [2, 9], 
    [4, 8], 
    [6, 7],
    [8, 6],
    [10, 5]
])

y = np.array([50, 65, 80, 90, 95])
```

```python
# Train the model
model = LinearRegression()
model.fit(X,y)

# coefficients
print("Intercept (bias) :", model.intercept_)
print("Coefficients :", model.coef_)
```

```python
# predictions for new students
new_data = np.array([
    [5, 8], 
    [7, 6], 
    [10, 5]
])
predictions = model.predict(new_data)

for features, pred in zip(new_data, predictions):
    print(f"Hours studied = {features[0]}, Hours slept={features[1]} => predicted score = {pred :2f}")
```
